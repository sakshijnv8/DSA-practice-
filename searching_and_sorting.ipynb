# **Linear search**

Definition


---


Linear search checks every element sequentially until the desired element is found.

Algorithm


---


1 Start from first element

2 Compare with target

3 If match → return index

4 Else move to next element

5 Continue until list ends

### **print with index**



```python
L=[10,30,23,48,56,16,1,3]
target=1

for i in range(len(L)):
  if L[i]==target:
    print(f"{target} found at {i}")
    break
else:
  print("not found")
```

**print without index**

```python
L=[10,30,23,48,56,16,1,3]
target=1

for i in L:
  if i == target:
    print(f"{target} found ")
    break
else:
  print("not found")
```

**using function**

```python
def linear_search(L, target):
  for i in range(len(L)):
    if L[i] == target:
      return i
  return -1
print(linear_search(L,48))
```

Time Complexity

Worst case -
O(n)

Because every element may be checked.

**Real Life Example**

Searching a name in an attendance list.
You check one name after another until you find it.

# **Binary Search**
Binary search works only on sorted arrays.
It repeatedly divides the search space into half.

**Steps**

1 Middle = 30

2 40 > 30 → search right half

3 Middle = 50

4 40 < 50 → search left half

5 Found


---


**Algorithm**

1 Find middle element

2 Compare with target

3 If equal → return

4 If smaller → search right  (#talking of mid)

5 If larger → search left

```python
def bs(a,t):
  l=0
  r=len(a)-1
  while l<=r:
    m=(l+r) // 2
    if a[m] ==t:
      return m
    elif a[m] < t:
      l=m+1
    else:
      r=m-1
  return -1
a=[1,3,5,7,9]
t=7
print(bs(a,t))
```

# **Lower bound**

agar array me target present ho to uska index ho jayega and if not then uske
aage vale element ka index le lega or jis postion pr adjust hoga uski index in a sorted array.

# code

```python
def lb(a,t):
  l=0
  r=len(a)-1
  ans=len(a)
  while l<=r:
    m=(l+r) // 2
    if a[m] >= t:
      ans=m
      r=m-1
    else:
      l=m+1
  return ans
a=[1,2,4,6,7,8,9,98]
t=4
print(lb(a,t))
```

# **Rotated array**
A rotated sorted array is an array that was originally sorted, but then shifted (rotated) at some point.

👉 Example:

Original sorted array:

[1, 2, 3, 4, 5, 6, 7]

After rotation:

[4, 5, 6, 7, 1, 2, 3]

👉 The point where it breaks (7 → 1) is called the pivot.

# code

```python
def rs(a,t):
  l=0
  r=len(a)-1
  while l<=r:
    m=(l+r) // 2
    if a[m] ==t:
      return m
    elif a[m]>=a[l]:
      if a[l] <= t < a[m]:
        r=m-1
      else:
        l=m+1
    else:
      if a[m] < t <= a[r]:
        l=m+1
      else:
        r=m-1
  return -1
a=[5,6,7,9,10,1,2,3]
t=3
print(rs(a,t))
```

# **Mountain array**
A mountain array (also called bitonic array) is:

👉 First strictly increasing

👉 Then strictly decreasing

**📌 Example**

arr = [1, 3, 5, 7, 6, 4, 2]

Increasing → 1, 3, 5, 7

Decreasing → 7, 6, 4, 2

Peak = 7 (index 3)

# code

```python
#from brute force
a=[1,2,7,3,2,0]
n=len(a)
for i in range (0,n):
  if (a[i] > a[i-1] and a[i] > a[i+1]):
    print((f"{i} is the index of peak element "))
```

time complexity:-
o(n)

```python
def ma(a):
    l = 0
    r = len(a) - 1
    while l < r:
        m = (l + r) // 2
        if a[m] < a[m+1]:
            l = m + 1
        else:
            r = m
    return l

a = [1,2,7,3,2,0]
print(ma(a))
```

time complexity:-
o(log(n))

# **Bubble Sort**
Bubble sort repeatedly compare elements and swaps adjacent elements if they are in the wrong order at the last index of array.



# Algorithm :

1.   start from index 0
2.   compare adjacent element
3.   swap if array of left> array of right
4.   repeat till end
5.   after each iteration largest element is fix.



**bubble sort in increasing order**

```python
def bubble_sort(a):       #take a function
    n = len(a)            #define n
    for i in range(0,n):  # loop to run from starting to end
      is_swap = False
      for j in range(0,n-i-1): # now if loop run it leaves the last one because it already place in last position in second time
        if a[j] > a[j+1]:         #sort in ascending order
          a[j],a[j+1] = a[j+1],a[j]
          is_swap=True
      if is_swap ==False:
        break                  #to stop loop iteration
    return a
a=[2,8,7,5,8,6,10]
print(bubble_sort(a))
```

**bubble sort in decreasing order**

```python
def bubble_sort(a):       #take a function
    n = len(a)            #define n
    for i in range(0,n):  # loop to run from starting to end
      is_swap = False
      for j in range(0,n-i-1): # now if loop run it leaves the last one because it already place in last position in second time
        if a[j] < a[j+1]:         #sort in descending order
          a[j],a[j+1] = a[j+1],a[j]
          is_swap=True
      if is_swap ==False:
        break                  #to stop loop iteration
    return a
a=[2,8,7,5,8,6,10]
print(bubble_sort(a))
```

# **Insertion Sort**
ek iteration me ek portion sort ho rha (ek element ko lega and compare with every element then insert at right position)

# Algorithm :

1.   start from second element
2.   compare with all previous elements
3.   shift element right if not sorted
4.   insert at correct position
insertion sort.png

**insertion sort in increasing order**

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]          # element to be placed
        j = i - 1             # compare with previous elements

        while j >= 0 and arr[j] > key:    # shift elements greater than key
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key   # place key at correct position

    return arr

arr=[5,32,7,13,10]
print(insertion_sort(arr))
```

**insertion sort in decreasing order**

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]          # element to be placed
        j = i - 1             # compare with previous elements

        while j >= 0 and arr[j] < key:    # shift elements smaller than key
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key   # place key at correct position

    return arr

arr=[5,32,7,13,10]
print(insertion_sort(arr))
```

# **Selection sort**
 isme hum sabse chote ko selct karenge then first index pr place kr dengethen baaki me sabse chhota then put after one.

## ALGORITHM :

1.   Start from index 0
2.   Assume current element is minimum
3.   Compare it with remaining elements
4.   Find the actual minimum
5.   Swap it with current position
6.   Move to next index
7.   Repeat until array is sorted
SELECTION.png

**increasing order of selection sort**

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i   # assume current element is smallest
        for j in range(i+1, n):   #Search in remaining array
            if arr[j] < arr[min_index]:  # Update smallest element
                min_index = j

        arr[i], arr[min_index] = arr[min_index], arr[i]  #SWAP
    return arr
arr=[5,32,7,13,10]
print(selection_sort(arr))
```

**decreasing order of selection sort**

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i   # assume current is smallest
        for j in range(i+1, n):
            if arr[j] > arr[min_index]:
                min_index = j

        arr[i], arr[min_index] = arr[min_index], arr[i]  #SWAP
    return arr
arr=[5,32,7,13,10]
print(selection_sort(arr))
```
