# **Power of number**

```python
def power(x,n):
  if n==0:
    return 1
  return x*power(x,n-1)
print(power(2,5))
```

# **Count digits of a number**

```python
def count_digits(n):
  if n < 10:
    return 1
  return 1 + count_digits(n//10)

count_digits(1354)
```

# **Print numbers from 1 to N**

```python
def print_numbers(n):
  if n==0:
    return
  print_numbers(n-1)
  print(n)

print_numbers(5)
```

# **Sum of digits**

```python
def sum(n):
  if n==0:
    return 0
  return n+sum(n-1)
print(sum(5))
```

# **Factorial of a number**

```python
def fact(n):
  if n==0 or n==1:
    return 1
  else:
    return n*fact(n-1)
print(fact(5))
```

# **Find fibnacci series**

```python
def fib(n):
  if n==0 or n==1:
    return n
  return fib(n-1)+fib(n-2)
for i in range(6):
  print(fib(i),end=',')
```

# **prime number using recursion**

```python
def prime(n,i):
  if n<=1:
    return False
  elif i*i>n:
    return True
  elif n%i==0:
    return False
  else:
    return prime(n,i+1)

print(prime(5,9))
```

# **Check palindrome**

```python
def is_palindrome(s):
  if len(s)<=1:
    return True
  if s[0] != s[-1]:
    return False
  return is_palindrome(s[1:-1])
print(is_palindrome("madam"))
```

# **reverse of a string**

```python
def rev(s):
  if s=="":
    return ""
  return rev(s[1:]) + s[0]
print(rev("sakshi"))
```

# **N QUEENS PROBLEM**
queens.jpg

**Backtracking Strategy**

General approach:

place queen in row if safe move to next row or else try next column

if stuck remove previous queen

**ALGORITHM**

1.   START
2.   Create an empty chessboard
3.   Start from row 0
4.   For each row in that row check if placing queen is safe
5.   If safe place queen
6.   Recursively place queen in next row
7.   If solution found - Print board
8.   If stuck - remove previous queen(backtracking)
9.   END




PSEUDOCODE

solve(row):
   
   if row == n
   
      print solution

      return

   for each column in board

      if possible is safe

         place queen

         solve(row+1)

         remove queen  (backtracking)

```python
def is_safe(board,row,col,n):
   #check column
   for i in range(row):
    if board[i][col] == 1:
      return False

   # Check left diagonal
   i= row-1
   j=col-1
   while i >=0 and j >= 0:
    if board[i][j]==1:
      return False
    i -= 1
    j -= 1

   #check right diagonal
   i= row - 1
   j= col + 1
   while i>= 0 and j< n:
    if board[i][j]== 1:
      return False
    i-= 1
    j+= 1
   return True

def solve_n_queens(board,row,n):
  if row == n:
    print_solution(board,n)
    return
  for col in range(n):

    if is_safe(board,row,col,n):
       board[row][col] = 1
       solve_n_queens(board,row+1,n)
       board[row][col] = 0   #backtrack

def print_solution(board,n):

  for i in range(n):
    for j in range(n):
      if board[i][j] == 1:
        print("Q" , end=" ")
      else:
        print(".", end=" ")
    print()
  print()

n= 4
board=[[0]*n for i in range(n)]

solve_n_queens(board,0,n)
```
