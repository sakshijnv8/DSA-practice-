# **LINKED LIST**

A linked list is a type of linear data structure in which each nodes are linked with each other through a connection.


*   A node contains two things first is data and second is a link that connects it with another node it means their address.
*   The first node is called the head node and we can traverse the whole list using this head and next links.


# **Types of Linked Lists**
There are three basic forms of linked lists:


1.   Singly linked lists
2.   Doubly linked lists
3.   Circular linked lists


# **NOTE :-**

Binary search is not possible for linked lists because the algorithm is based on jumping directly to different array elements, and that is not possible with linked lists.

# **These are some key linked list properties, compared to arrays:**

* Linked lists are not allocated to a fixed size in memory like arrays are, so linked lists do not require to move the whole list into a larger memory space when the fixed memory space fills up, like arrays must.

* Linked list nodes are not laid out one right after the other in memory (contiguously), so linked list nodes do not have to be shifted up or down in memory when nodes are inserted or deleted.

* Linked list nodes require more memory to store one or more links to other nodes. Array elements do not require that much memory, because array elements do not contain links to other elements.

* Linked list operations are usually harder to program and require more lines than similar array operations, because programming languages have better built in support for arrays.

* We must traverse a linked list to find a node at a specific position, but with arrays we can access an element directly by writing myArray[5].

# TRAVERSE IN LINKED LIST

Traversing a linked list means to go through the linked list by following the links from one node to the next.

To traverse a singly linked list, we start with the first node in the list, the head node, and follow that node's next link, and the next node's next link and so on, until the next address is null.

```python
# traverse in linked list

def printlinkedlist(head):    #a function
    curr = head               #starting from first node
    while curr!= None:        #traverse through last node i.e iterate till reach the last node
        print(curr.data , end=' -> ')  # print the data
        curr = curr.next              #move to next node after every iteration
printlinkedlist(head)                # call the function and print the list finally
```

# INSERTION OF NODE AT BEGINNING

To insert a node in a linked list we first need to create the node, and then at the position where we insert it, we need to adjust the pointers so that the previous node points to the new node, and the new node points to the correct next node.

```python
# insertion in a linked list
# insertion at the begining

new_node = Node(1)
new_node.next = head
head=new_node
print(head.data)
print("Now print the linked list")
printlinkedlist(head)
```

# INSERTION OF NODE AT THE END

```python
# insertion at the end
new_node = Node(8)
def insert_end(head):
    curr = head
    while curr.next!= None:
        curr = curr.next
    curr.next=new_node
    new_node.next = None
insert_end(head)
printlinkedlist(head)
```

So before deleting the node, we need to get the next pointer from the previous node, and connect the previous node to the new next node before deleting the node in between.

```python
# Deletion in linked list
# deletion of the first node

print("before deletetion\n")
printlinkedlist(head)

head = head.next           #just made the second node as head then the first one is already diconnected from the list

print("\nafter deletion\n")
printlinkedlist(head)
```

# **DOUBLY LINKED LIST**

A doubly linked list has nodes with addresses to both the previous and the next node, like in the image below, and therefore takes up more memory. But doubly linked lists are good if you want to be able to move both up and down in the list.

```python
# DOUBLY LINKED LIST

class DoublyLL:
    def __init__ (self,value):
        self.data=value
        self.next=None
        self.prev=None

a = DoublyLL(3)
b = DoublyLL(4)
c = DoublyLL(7)

a.next = b
b.prev = a
b.next = c
c.prev = b

print(a.data)
print(a.next.data)
print(b.prev.data)
print(b.next.data)
print(c.prev.data)
```

# **CIRCULAR LINKED LIST**

A circular linked list is like a singly or doubly linked list with the first node, the "head", and the last node, the "tail", connected.

In singly or doubly linked lists, we can find the start and end of a list by just checking if the links are null. But for circular linked lists, more complex code is needed to explicitly check for start and end nodes in certain applications.

Circular linked lists are good for lists you need to cycle through continuously.

image.png

```python
#CIRCULAR LINKED LIST


class circularlist:
  def __init__(self, data):
    self.data = data
    self.next = None

a = circularlist(5)
b = circularlist(8)
c = circularlist(10)

a.next = b
b.next = c
c.next = a

head = a

curr = head

while True:
  print(curr.data, end = " -> ")
  curr = curr.next

  if curr == head:
    break


```

# CALCULATE LENGTH AND FIND MID VALUE IN A LINKED LIST

# **SORTING OF LINKED LIST USING MERGE SORT**

```python
#sorting linked list using merge sort

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def sortList(self, head):
        # Base case: empty list or single node (already sorted)
        if not head or not head.next:
            return head

        # Step 1: Find middle of list
        slow, fast = head, head.next
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        mid = slow.next
        slow.next = None  # Split list into two halves

        # Step 2: Recursively sort both halves
        left = self.sortList(head)
        right = self.sortList(mid)

        # Step 3: Merge sorted halves
        return self.merge(left, right)

    def merge(self, l1, l2):
        dummy = ListNode()
        tail = dummy

        while l1 and l2:
            if l1.val < l2.val:
                tail.next = l1
                l1 = l1.next
            else:
                tail.next = l2
                l2 = l2.next
            tail = tail.next

        tail.next = l1 if l1 else l2
        return dummy.next
curr = head

while curr:
    print(curr.data, end=" -> ")
    curr = curr.next

```

# **REVERSE OF LINKED LIST**

```python
#reverse of linked list

head = a

def rev(head):
  curr = head
  nxt = prev = None    #store in a temporary variable

  while curr != None:
    nxt = curr.next
    curr.next = prev
    prev = curr
    curr = nxt

  return prev # the head of the reversed linked list

print("Linked List Before Reversal:")
printlinkedlist(head)

head = rev(head)

print("\nLinked List After Reversal:")
printlinkedlist(head)
```

# **ROTATION OF LINKED LIST**

```python
#ROTATION OF A LINKED LIST

class ListNode:     #CREATE A LINKED LIST
    def __init__(self, val):
        self.val = val
        self.next = None

def printList(head):

    curr = head
    while curr:
        print(curr.val, end=" -> ")      #print list
        curr = curr.next

def rotateRight(head, k):

    if not head or not head.next or k == 0:
        return head

    # Step 1: find length
    length = 1
    tail = head

    while tail.next:
        tail = tail.next
        length += 1

    # Step 2: make circular linked list
    tail.next = head

    # Step 3: reduce k if larger than length
    k = k % length

    # Step 4: find new tail
    steps = length - k

    newTail = head

    for i in range(steps - 1):
        newTail = newTail.next

    # Step 5: new head
    newHead = newTail.next

    # break the circle
    newTail.next = None

    return newHead


# creating linked list
a = ListNode(1)
b = ListNode(2)
c = ListNode(3)
d = ListNode(4)
e = ListNode(5)

a.next = b
b.next = c
c.next = d
d.next = e

head = a

print("Before Rotation:")
printList(head)

head = rotateRight(head, 2)

print("\nAfter Rotation:")
printList(head)
```
